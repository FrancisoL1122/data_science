```python
import matplotlib.pyplot as plt
```

```python
labels = ['Frogs','Dogs','Cats','Birds']

sizes = [15,30,45,10]

explode =(0,0.1,0,0)

fig1,axl = plt.subplots()
axl.pie(sizes, explode=explode, labels =labels, autopct='%1.1f%%',shadow=True, startangle=90)
axl.axis('equal');
```
